```python
import cv2 as cv
from google.colab.patches import cv2_imshow

# Make these global variables so they can be accessed when imported
image = None
greyscaled_image = None
binary_thresh = None
adaptive_thresh_mean = None
adaptive_thresh_guass = None


def load_and_process_image():
    global image, greyscaled_image, binary_thresh, adaptive_thresh_mean, adaptive_thresh_guass

    image = cv.imread('/content/sample.png')

    # Check if image was loaded successfully
    if image is None:
        print("Error: Could not load image. Check if the file exists and is a valid image format.")
        return False

    # Convert to greyscale
    greyscaled_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

    # Apply binary threshold
    _, binary_thresh = cv.threshold(greyscaled_image, 150, 255, cv.THRESH_BINARY)

    # Apply adaptive thresholds
    adaptive_thresh_mean = cv.adaptiveThreshold(
        greyscaled_image, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 5)
    adaptive_thresh_guass = cv.adaptiveThreshold(
        greyscaled_image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 12)

    return True


def binary_threshold():
    if binary_thresh is not None:
        cv2_imshow(binary_thresh)


def adaptive_mean_threshold():
    if adaptive_thresh_mean is not None:
        cv2_imshow(adaptive_thresh_mean)


def adaptive_guassian_threshold():
    if adaptive_thresh_guass is not None:
        cv2_imshow(adaptive_thresh_guass)


def main():
    if load_and_process_image():
        print("Binary threshold:")
        binary_threshold()
        print("Adaptive mean threshold:")
        adaptive_mean_threshold()
        print("Adaptive Gaussian threshold:")
        adaptive_guassian_threshold()


if __name__ == "__main__":
    main()
```
